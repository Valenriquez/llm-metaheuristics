```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Use the flat_data format (you can copy the data from above)
flat_data = np.array(f_best_values)
df = pd.DataFrame(flat_data, columns=['Step', 'Value'])

# Scale up the values to match the original scale
df['Value'] = df['Value'] * 1000

# Create the plot
plt.figure(figsize=(12, 8))

# Create violin plot
sns.violinplot(data=df, x='Step', y='Value', color='lightgreen', alpha=0.5)

# Add performance line
performance_means = df.groupby('Step')['Value'].mean()
plt.plot(range(len(performance_means)), performance_means, 'o--', color='orange', label='Performance')

# Customize the plot
plt.xlabel('Step')
plt.ylabel('Fitness')
plt.title('Optimization Performance Over 10 Steps')
plt.legend(['Performance', 'Raw data'])

# Add grid
plt.grid(True, alpha=0.3)

# Adjust y-axis limits
plt.ylim(0, 100000)

# Show plot
plt.show()

# Print summary statistics
print("\nSummary Statistics:")
for step in range(10):
    step_data = df[df['Step'] == step]['Value']
    print(f"\nStep {step}:")
    print(f"Number of values: {len(step_data)}")
    print(f"Mean: {step_data.mean():.2f}")
    print(f"Std: {step_data.std():.2f}")
```
